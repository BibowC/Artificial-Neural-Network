```python
#FORMULAS DEF (VEL, REYNOLDS, F)

def Reynolds2(mfr, dh, acs, mu):
    Re = (mfr * dh) / (acs * mu)
    return Re 

def Velocity(mfr, rho, acs):
    Vel = (mfr) / (rho * acs)
    return Vel

def ffactor(dp, dh, L, rho, V):
    f = (dp * dh) / (2*L*rho*(np.power(V,2)))
    return f
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
import plotly.graph_objects as go
from scipy.interpolate import griddata

# Load your dataset
# ANNDF = pd.read_csv('your_dataset.csv')  # Replace with your actual file path

# Shuffle the dataset
ANNDF = shuffle(ANNDF)

# Select inputs (X) and outputs (Y)
X = ANNDF.iloc[:, [0, 2]].values  # Assuming columns 0 and 2 are Reynolds and Velocity
Y = ANNDF.iloc[:, 1].values.reshape(-1, 1)  # Assuming column 1 is the output

# Normalize the data
scalerx = StandardScaler()
scalery = StandardScaler()
X_scaled = scalerx.fit_transform(X)
Y_scaled = scalery.fit_transform(Y)

# Define K-Fold Cross-Validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Define the ANN model
mlp = MLPRegressor(max_iter=1000)

# Define a grid of hyperparameters to search over
parameter_space = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],  # Example configurations
    'activation': ['tanh', 'relu'],
    'solver': ['adam', 'sgd'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant', 'adaptive'],
}

# Apply GridSearchCV to find the best parameters
grid_search = GridSearchCV(mlp, parameter_space, n_jobs=-1, cv=kf, scoring='neg_mean_squared_error')
grid_search.fit(X_scaled, Y_scaled.ravel())

# Get the best model
best_mlp = grid_search.best_estimator_
print(f"Best Parameters: {grid_search.best_params_}")

# Predict using the best model
Y_pred_scaled = best_mlp.predict(X_scaled)
Y_pred = scalery.inverse_transform(Y_pred_scaled.reshape(-1, 1))

# Calculate Mean Squared Error
mse = mean_squared_error(scalery.inverse_transform(Y_scaled), Y_pred)
print(f"Mean Squared Error: {mse}")

# Prepare data for Plotly
X_plot = scalerx.inverse_transform(X_scaled)
X1 = X_plot[:, 0]
X2 = X_plot[:, 1]
Y_true = scalery.inverse_transform(Y_scaled).flatten()
Y_pred = Y_pred.flatten()

# Create grid for plotting
grid_x, grid_y = np.mgrid[X1.min():X1.max():100j, X2.min():X2.max():100j]
grid_z = griddata((X1, X2), Y_pred, (grid_x, grid_y), method='linear')

# Create a 3D surface plot using Plotly
fig = go.Figure()

# Add true values scatter plot
fig.add_trace(go.Scatter3d(
    x=X1, y=X2, z=Y_true,
    mode='markers',
    marker=dict(size=5, color='blue', opacity=0.8),
    name='True Values'
))


fig.add_trace(go.Scatter3d(
    x=Recor1, y=stpkyan, z=cor1(Recor1),
    mode='markers',
    marker=dict(size=5, color='red', opacity=0.8),
    name='Kaixin Yan/As Built'
))

fig.add_trace(go.Scatter3d(
    x=Recor1, y=stpkyan, z=cor2(Recor1),
    mode='markers',
    marker=dict(size=5, color='green', opacity=0.8),
    name='Kaixin Yan/As Polished'
))


# Add predicted surface plot
fig.add_trace(go.Surface(
    x=grid_x, y=grid_y, z=grid_z,
    colorscale='Viridis',
    opacity=0.7,
    name='Predicted Surface'
))

fig.update_layout(
    title='3D Surface Plot',
    scene=dict(
        xaxis_title='Reynolds Number',
        yaxis_title='Standard Porosity',
        zaxis_title='f factor'
    )
)

```
