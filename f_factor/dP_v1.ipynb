```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo 
import CoolProp.CoolProp as CP


# CS4_V1 VALUES
Lf = 120/1000 # mm Length of the core
Vf_cs4 = 13190 * 1e-9 #m³  Volume of the fluid in core
Afs_cs4 = 35459 * 1e-6 #m² Interfacial surface area between fluid/wall
Acs_cs4 = Vf_cs4 / Lf
Dh_cs4 = (4*Vf_cs4)/Afs_cs4

#CS6_V1 VALUES
Lf = 120/1000 # mm Length of the core
Vf_cs6 = 17310 * 1e-9 #m³  Volume of the fluid in core
Afs_cs6 = 24953 * 1e-6 #m² Interfacial surface area between fluid/wall
Acs_cs6 = Vf_cs6 / Lf
Dh_cs6 = (4*Vf_cs6)/Afs_cs6

#CS8_V1 VALUES
Lf = 120/1000 # m Length of the core
Vf_cs8 = 19478 * 1e-9 #m³  Volume of the fluid in core
Afs_cs8 = 18843 * 1e-6 #m² Interfacial surface area between fluid/wall
Acs_cs8 = Vf_cs8 / Lf
Dh_cs8 = (4*Vf_cs8)/Afs_cs8

#CS10_V1 VALUES
Lf = 120/1000 # m Length of the core
Vf_cs10 = 20892 * 1e-9 #m³  Volume of the fluid in core
Afs_cs10 = 15415 * 1e-6 #m² Interfacial surface area between fluid/wall
Acs_cs10 = Vf_cs10 / Lf
Dh_cs10 = (4*Vf_cs10)/Afs_cs10

#CS12_V1 VALUES
Lf = 120/1000 # m Length of the core
Vf_cs12 = 21921 * 1e-9 #m³  Volume of the fluid in core
Afs_cs12 = 12935 * 1e-6 #m² Interfacial surface area between fluid/wall
Acs_cs12 = Vf_cs12 / Lf
Dh_cs12 = (4*Vf_cs12)/Afs_cs12

#CS15_V1 VALUES
Lf = 120/1000 # m Length of the core
Vf_cs15 = 22417 * 1e-9 #m³  Volume of the fluid in core
Afs_cs15 = 10438 * 1e-6 #m² Interfacial surface area between fluid/wall
Acs_cs15 = Vf_cs15 / Lf
Dh_cs15 = (4*Vf_cs15)/Afs_cs15

# DEFINING REYNOLDS AND VELOCITY
T = 300
mu = CP.PropsSI('V', 'T', T, 'P', 101325, 'Water')
rho = CP.PropsSI('D', 'T', 300, 'P', 101325, 'Water')


def Reynolds(mfr, dh, acs, mu):
    Re = (mfr * dh) / (acs * mu)
    return Re 

def Reynolds(rho, v, dh, mu):
    Re = (rho * v * dh) / (mu)
    return Re 

def Velocity(mfr, rho, acs):
    Vel = (mfr) / (rho * acs)
    return Vel

def ffactor(dp, dh, L, rho, V):
    f = (dp * dh) / (2*L*rho*(np.power(V,2)))
    return f

```

```python
#CS4 DATA
col = [9, 8, 10]


df_cs4 = []

for freq in range(30, 55, 5):
    sheetname = f'{freq}'
    df = pd.read_excel('cs4_v1.xlsx', sheetname)
    df = df.iloc[:, col].values
    df_cs4.append(df)

mfr_cs4 = []
dp_cs4 = []
abs_cs4 = []

for df in df_cs4:
    massflowrate_cs4 = df[:, 0]
    differencial_cs4 = df[:, 1] * 100000
    absolute_cs4 = df[:, 2]
    lastmfr_cs4 = massflowrate_cs4[-100:]
    lastdp_cs4 = differencial_cs4[-100:]
    lastabs_cs4 = absolute_cs4[-100:]        
    
    meanmfr_cs4 = lastmfr_cs4.mean()
    meandp_cs4 = lastdp_cs4.mean()
    meanabs_cs4 = lastabs_cs4.mean() 

    mfr_cs4.append(meanmfr_cs4)
    dp_cs4.append(meandp_cs4)
    abs_cs4.append(meanabs_cs4)

```

```python

# REYNOLDS AND VELOCITY VALUES

mfr_cs4 = np.array(mfr_cs4)
mfr_cs6 = np.array(mfr_cs6)
mfr_cs8 = np.array(mfr_cs8)
mfr_cs10 = np.array(mfr_cs10)
mfr_cs12 = np.array(mfr_cs12)
mfr_cs15 = np.array(mfr_cs15)

V_cs4 = np.array(Velocity(mfr_cs4, rho, Acs_cs4))
V_cs6 = np.array(Velocity(mfr_cs6, rho, Acs_cs6))
V_cs8 = np.array(Velocity(mfr_cs8, rho, Acs_cs8))
V_cs10 = np.array(Velocity(mfr_cs10, rho, Acs_cs10))
V_cs12 = np.array(Velocity(mfr_cs12, rho, Acs_cs12))
V_cs15 = np.array(Velocity(mfr_cs15, rho, Acs_cs15))

Re_cs4_v1 = Reynolds(1000, V_cs4, Dh_cs4, mu)
Re_cs6_v1 = Reynolds(1000, V_cs6, Dh_cs6, mu)
Re_cs8_v1 = Reynolds(1000, V_cs8, Dh_cs8, mu)
Re_cs10_v1 = Reynolds(1000, V_cs10, Dh_cs10, mu)
Re_cs12_v1 = Reynolds(1000, V_cs12, Dh_cs12, mu)
Re_cs15_v1 = Reynolds(1000, V_cs15, Dh_cs15, mu)


f_cs4 = ffactor(np.array(dp_cs4), Dh_cs4, Lf, rho, V_cs4)
f_cs6 = ffactor(np.array(dp_cs6), Dh_cs6, Lf, rho, V_cs6)
f_cs8 = ffactor(np.array(dp_cs8), Dh_cs8, Lf, rho, V_cs8)
f_cs10 = ffactor(np.array(dp_cs10), Dh_cs10, Lf, rho, V_cs10)
f_cs12 = ffactor(np.array(dp_cs12), Dh_cs12, Lf, rho, V_cs12)
f_cs15 = ffactor(np.array(dp_cs15), Dh_cs15, Lf, rho, V_cs15)
```
