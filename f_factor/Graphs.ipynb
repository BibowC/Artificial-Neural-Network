```python
#FORMULAS DEF (VEL, REYNOLDS, F)

def Reynolds2(mfr, dh, acs, mu):
    Re = (mfr * dh) / (acs * mu)
    return Re 

def Velocity(mfr, rho, acs):
    Vel = (mfr) / (rho * acs)
    return Vel

def ffactor(dp, dh, L, rho, V):
    f = (dp * dh) / (2*L*rho*(np.power(V,2)))
    return f
```

```python
def Traces(x, y, name, color, symbol): #symbol = circle, square, diamond, cross, triangle-up, pentaon, hexagon, star
    trace = go.Scatter(
        x=x,
        y=y,
        name=name,
        mode='markers',
        marker=dict(
            size=15,
            color=color,
            symbol=symbol,  # Forma do marcador
            line=dict(
                width=2,
                color='Black'  # Cor da borda do marcador
            )
        )
    )
    return trace



def curves(trace, title, ytitle):

    fig = go.Figure(data = trace)

    fig.update_layout(
    font=dict(
    size=25),
    title={
        'text': title,
        'x': 0.5,
        'xanchor': 'center'
    },
    xaxis={
        'title': 'Reynolds',
        'showline': True,
        'showgrid': False,
        'zeroline': True,
        'zerolinecolor': 'black',
        'linecolor': 'black',
        'ticks': 'outside',
        'mirror': True
    },
    yaxis={
        'title': ytitle,
        'showline': True,
        'showgrid': False,
        'zeroline': True,
        'zerolinecolor': 'black',
        'linecolor': 'black',
        'ticks': 'outside',
        'mirror': True
    },
    plot_bgcolor='white',
    hovermode='closest'
)
    
    pyo.plot(fig)
```
